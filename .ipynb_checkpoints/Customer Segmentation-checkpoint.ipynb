```python
#imports and dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
#ignore warnings
warnings.filterwarnings('ignore')
```

```python
#read csv file
hotelDF = pd.read_csv('clean_hotel_second.csv')
hotelDF
```

```python
#select the features for clustering
features = hotelDF[['lead_time','total_nights', 'total_price','no_of_adults','no_of_children', 'repeated_guest']].values
```

```python
#standardize features
stanScaler = StandardScaler()
scaledFeatures = stanScaler.fit_transform(features)
```

```python
#calculate inertia of each cluster number 
#and graph the elbow method
clusterRange = range(1,11)

inertia = []

for k in clusterRange:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(8, 5))
plt.plot(clusterRange, inertia, 'bo-', markersize=8)
plt.xlabel('Cluster Numbers')
plt.ylabel('Inertia')
plt.title('Elbow Method to Choose Cluster Number')
plt.show()
```

```python

# Apply PCA to reduce dimensions
pca = PCA(n_components=2)  # You can adjust the number of components based on your analysis
principalComponents = pca.fit_transform(scaledFeatures)

# Optionally, check explained variance ratio to decide on components
print(pca.explained_variance_ratio_)

# You can use the principal components for clustering instead of the scaled features
```

```python

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Apply KMeans with a range of cluster sizes to use the elbow method
inertia = []
silhouette_scores = []
cluster_range = range(2, 11)

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(principalComponents)
    
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(principalComponents, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Plotting the elbow method
plt.figure(figsize=(10, 5))
plt.plot(cluster_range, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal Clusters')
plt.show()

# Plotting the silhouette scores
plt.figure(figsize=(10, 5))
plt.plot(cluster_range, silhouette_scores, 'ro-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores For Various Clusters')
plt.show()
```

```python
# KMeans cluster with four clusters
kMeans = KMeans(n_clusters=4, random_state=42)
hotelDF['cluster'] = kmeans.fit_predict(scaledFeatures)
```

```python
#view first five rows of dataframe
hotelDF[['lead_time','total_nights', 'total_price','no_of_adults','no_of_children', 'repeated_guest', 'cluster']].head()
```

```python
#k-clustering of leadtime vs total price
x_feature = 'lead_time'
y_feature = 'total_price'

# Plot the clusters
plt.figure(figsize=(10, 6))
plt.scatter(hotelDF[x_feature], hotelDF[y_feature], c=hotelDF['cluster'], cmap='rainbow', marker='o', edgecolor='k')
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.title('K-means Clustering (k=4)')
plt.colorbar(label='Cluster Label')
plt.show()
```
